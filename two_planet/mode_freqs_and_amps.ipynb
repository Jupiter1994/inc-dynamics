# Mode frequencies and amplitudes for the 2-planet case

We compare the numerical and analytical mode frequencies and amplitudes for various values of $\alpha \equiv a_i / a_{i+1}$.  

Mode (angular) frequency is denoted as $\omega$ or $g$, and the mode amplitude eigenvector for the 2-planet case is denoted as $\vec{v}_1 = 
\begin{pmatrix}
v_1^B \\
v_1^C
\end{pmatrix}
$. $v_1^B$ is the component of $\vec{v}_1$ corresponding to the innermost planet (or Planet B), and $v_1^C$ corresponds to the outermost planet (or Planet C).

```python
# import packages
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 7)
```

Define the arrays that contain the derived numerical and analytical values of $\omega$ and $\vec{v}_1$.

```python
# define the arrays containing our results
alphas = np.asarray([1/2,1/1.9,1/1.8,1/1.7,1/1.6,1/1.5,1/1.4,1/1.3])

## numerical results
omegas_num = np.asarray([2.12E-05,2.67E-05,3.44E-05,4.57E-05,\
                         6.26E-05,9.15E-05,1.42E-04,2.49E-04])
v1_B_num = np.asarray([3.64005568,3.60106005,3.5598378,3.51612108,\
                       3.47025645,3.41950506,3.36846165,3.30741692])
v1_C_num = np.asarray([2.59158848,2.63049346,2.6717068,2.71547592,\
                       2.7626021,2.81158199,2.86687005,2.92133963])

## analytical results
omegas_an = np.asarray([2.17E-05,2.74E-05,3.55E-05,4.74E-05,\
                        6.59E-05,9.69E-05,1.54E-04,2.78E-04])
v1_B_an = np.asarray([3.63953082,3.60060901,3.56598655,3.52192884,\
                      3.47495806,3.4194275,3.36610192,3.30866125])
v1_C_an = np.asarray([2.59209608,2.63099399,2.6770972,2.72067589,\
                      2.76700707,2.81208513,2.86539129,2.92281496])
```

## Mode frequencies

```python
fig = plt.figure(figsize=(12,6))
fig.patch.set_facecolor('white')

plt.scatter(alphas, omegas_an, s=100, color='red', marker='o', label='Analytical result')
plt.scatter(alphas, omegas_num, s=100, color='blue', marker='P', label='Numerical result')

plt.xlabel('α')
plt.ylabel('Mode frequency (yr$^{-1}$)')
plt.title('Mode frequency ($\omega$) vs. α')
plt.legend()

fig_path = "Figures/2p_mode_freq.png"
plt.savefig(fig_path, bbox_inches='tight')
plt.show()
```

## Mode amplitudes

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12,12))
fig.patch.set_facecolor('white')

## Mode 1 amplitude for Planet B
ax1.scatter(alphas, v1_B_an, s=100, color='red', marker='o', label='Analytical result')
ax1.scatter(alphas, v1_B_num, s=100, color='blue', marker='P', label='Numerical result')

#ax1.set_xlabel('α')
ax1.set_ylabel('Mode 1 amplitude for Planet B (deg)')
ax1.set_title('Mode 1 amplitude for Planet B vs. α')
ax1.legend()

## Mode 1 amplitude for Planet C
ax2.scatter(alphas, v1_C_an, s=100, color='red', marker='o', label='Analytical result')
ax2.scatter(alphas, v1_C_num, s=100, color='blue', marker='P', label='Numerical result')

ax2.set_xlabel('α')
ax2.set_ylabel('Mode 1 amplitude for Planet C (deg)')
ax2.set_title('Mode 1 amplitude for Planet C vs. α')
ax2.legend()

fig_path = "Figures/2p_mode_amps.png"
plt.savefig(fig_path, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.scatter(alphas, v1_B_an, s=100, color='red', marker='o', label='Analytical')
plt.scatter(alphas, v1_B_num, s=100, color='blue', marker='P', label='Numerical')

plt.xlabel('alpha')
plt.ylabel('Mode 1 amplitude for Planet B,  ($v_1^B$)')
plt.title('Mode 1 amplitude for Planet B vs. alpha')

plt.legend()
plt.show()
```
