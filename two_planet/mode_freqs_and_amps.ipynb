# Mode frequencies and amplitudes for the 2-planet case

We compare the numerical and analytical mode frequencies and amplitudes for various values of $\alpha \equiv a_i / a_{i+1}$.  

Mode (angular) frequency is denoted as $\omega$ or $g$, and the mode amplitude eigenvector for the 2-planet case is denoted as $\vec{v}_1 = 
\begin{pmatrix}
v_1^B \\
v_1^C
\end{pmatrix}
$. $v_1^B$ is the component of $\vec{v}_1$ corresponding to the innermost planet (or Planet B), and $v_1^C$ corresponds to the outermost planet (or Planet C).

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt

## set global figure size
plt.rcParams["figure.figsize"] = (13, 10)
## set global font size
plt.rcParams.update({'font.size':18})
```

Define the arrays that contain the derived numerical and analytical values of $\omega$ and $\vec{v}_1$.

```python
# define the arrays containing our results
alphas = np.asarray([1/2,1/1.9,1/1.8,1/1.7,1/1.6,1/1.5,1/1.4,1/1.3])

## numerical results
omegas_num = np.asarray([2.12E-05,2.67E-05,3.44E-05,4.57E-05,\
                         6.26E-05,9.15E-05,1.42E-04,2.49E-04])
v1_B_num = np.asarray([3.64005568,3.60106005,3.5598378,3.51612108,\
                       3.47025645,3.41950506,3.36846165,3.30741692])
v1_C_num = np.asarray([2.59158848,2.63049346,2.6717068,2.71547592,\
                       2.7626021,2.81158199,2.86687005,2.92133963])

## analytical results
omegas_an = np.asarray([2.17E-05,2.74E-05,3.55E-05,4.74E-05,\
                        6.59E-05,9.69E-05,1.54E-04,2.78E-04])
v1_B_an = np.asarray([3.63953082,3.60060901,3.56598655,3.52192884,\
                      3.47495806,3.4194275,3.36610192,3.30866125])
v1_C_an = np.asarray([2.59209608,2.63099399,2.6770972,2.72067589,\
                      2.76700707,2.81208513,2.86539129,2.92281496])
```

## Mode frequencies

```python
fig = plt.figure(figsize=(14,8))
fig.patch.set_facecolor('white')

plt.scatter(alphas, np.log10(omegas_num), s=100, \
            marker='o', label='$g_1$ (numerical)')
plt.scatter(alphas, np.log10(omegas_an), s=100, facecolors='none', \
            edgecolors='black', label='$g_1$ (analytical)')


plt.xlabel('α')
plt.ylabel('$\log_{10}(g_1)$ (yr$^{-1}$)')
plt.title('Mode 1 frequency ($g_1$) vs. α (2-planet case)')
plt.legend()

fig_path = "Figures/2p_mode_freq.png"
plt.savefig(fig_path, bbox_inches='tight')
plt.show()
```

## Mode amplitudes

## Plotting numerically-derived components of a mode

For each numerically-derived mode amplitude vector $\vec{A}_j$, plot all of its components.

```python
fig = plt.figure(figsize=(16,8))
fig.patch.set_facecolor('white')

plt.scatter(alphas, v1_B_an, s=100, \
            marker='o', alpha=0.7, label='$A_{B1}$ (numerical)')
plt.scatter(alphas, v1_C_an, s=100, \
            marker='o', alpha=0.7, label='$A_{C1}$ (numerical)')

plt.xlabel('α')
plt.ylabel('Amplitude (deg)')
plt.ylim(0, 4)
plt.title('Numerically-derived mode 1 amplitudes (2-planet case)')
plt.legend()

fig_path = "Figures/2p_mode_amps_num.png"
plt.savefig(fig_path, bbox_inches='tight')
plt.show()
```

## Plotting relative differences between analytical and numerical amplitude components

```python
rel_amp_diff_B = v1_B_an - v1_B_num
rel_amp_diff_C = v1_C_an - v1_C_num
```

```python
fig = plt.figure(figsize=(16,8))
fig.patch.set_facecolor('white')

plt.scatter(alphas, rel_amp_diff_B, s=100, marker='x', \
            label='$A_{B1}$ difference')
plt.scatter(alphas, rel_amp_diff_C, s=100, marker='x', \
            label='$A_{C1}$ difference')

plt.xlabel('α')
plt.ylabel('Amplitude (deg)')
plt.ticklabel_format(axis='y', style='sci', scilimits=(-2,-2))

plt.ylim(-1e-2, 1e-2)
plt.title('Difference between analytical and numerical mode 1 amplitudes')
plt.legend()

fig_path = "Figures/2p_mode1_amp_diffs.png"
plt.savefig(fig_path, bbox_inches='tight')
plt.show()
```
