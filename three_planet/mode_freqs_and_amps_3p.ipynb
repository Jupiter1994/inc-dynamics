# Mode frequencies and amplitudes for the 3-planet case

We compare the numerical and analytical mode frequencies and amplitudes for various values of $\alpha \equiv a_i / a_{i+1}$.  

Mode (angular) frequency is denoted as $\omega$ or $g$, and the $j$-th mode amplitude eigenvector for the 3-planet case is denoted as $\vec{v}_j = 
\begin{pmatrix}
v_{Bj} \\
v_{Cj} \\
v_{Dj}
\end{pmatrix}
$. $v_{Bj}$ is the component of $\vec{v}_j$ corresponding to the innermost planet (or Planet B), $v_{Cj}$ corresponds to the middle planet (or Planet C), and $v_{Dj}$ corresponds to the outermost planet (or Planet D). For three planets, there are two mode amplitudes, so $j=1,2$.

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt

## set global figure size
plt.rcParams["figure.figsize"] = (10, 7)
## set global font size
plt.rcParams.update({'font.size':18})
```

Define the arrays that contain the derived numerical and analytical values of $\omega$ and $\vec{v}_1$.

```python
# define the arrays containing our results
alphas = np.asarray([1/2,1/1.9,1/1.8,1/1.7,1/1.6,1/1.5,1/1.4,1/1.3])

## numerical results
# mode freqs
g1_num = np.asarray([2.30E-05,2.93E-05,3.83E-05,5.17E-05,\
                      7.22E-05,1.08E-04,1.72E-04,3.13E-04])
g2_num = np.asarray([5.94E-06,8.11E-06,1.14E-05,1.65E-05,\
                      2.48E-05,3.98E-05,6.81E-05,1.31E-04])

# mode 1 amp vector: each column corresponds to the 3-component vector for
# a specific alpha
v1_num = np.asarray([[3.672868452,3.574495609,3.463909821,3.338612237,\
                      3.195234836,3.020547186,2.807025291,2.559005673],
                     [3.459243925,3.507712903,3.556377601,3.604155164,
                      3.648826149,3.683203845,3.698515431,3.685989327],
                     [0.414154556,0.4601997304,0.5148103561,0.5799174539,\
                      0.6578844696,0.7545125257,0.8729654186,1.00878603]])
# mode 2 amp vector
v2_num = np.asarray([[0.9663632955,1.027172197,1.098189956,1.182228698,\
                      1.282562395,1.410852508,1.566803674,1.784673956],
                     [0.6185092045,0.6278031668,0.6339387309,0.6347642827,\
                      0.6334150389,0.6159592268,0.5846009529,0.5159390037],
                     [0.8902795542,0.9648140052,1.050508342,1.149604324,\
                      1.269469553,1.411387147,1.583116834,1.800347417]])

## analytical results
# mode freqs
g1_an = np.asarray([2.36E-05,3.02E-05,3.97E-05,5.39E-05,\
                     7.63E-05,1.15E-04,1.87E-04,3.48E-04])
g2_an = np.asarray([6.07E-06,8.31E-06,1.17E-05,1.71E-05,\
                    2.61E-05,4.22E-05,7.42E-05,1.48E-04])

# mode 1 amp vector
v1_an = np.asarray([[3.81497005,3.71555396,3.60335265,3.47519939,\
                    3.32676132,3.15207153,2.94299577,2.68895835],
                   [3.57852003,3.62926144,3.67972853,3.72862201,\
                    3.77369064,3.81100112,3.83364172,3.82954467],
                    [0.42059171,0.46708082,0.52202415,0.5876412,\
                     0.66682614,0.76325497,0.88127542,1.02509661]])
# mode 2 amp vector
v2_an = np.asarray([[1.375696,1.40065649,1.43303642,1.475263,
                     1.53070694,1.60404811,1.701678,1.8318755],
                    [0.89858151,0.876983,0.85247527,0.82379871,
                     0.7889017,0.74436746,0.68441947,0.59929087],
                    [1.28085252,1.33132283,1.38992089,1.45880162,
                     1.54079634,1.63955985,1.75958491,1.90574163]])
```

## Mode frequencies

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(14,14))
fig.patch.set_facecolor('white')

ax1.scatter(alphas, np.log10(g1_num), s=100, \
            marker='o', label='$g_1$ (numerical)')
ax1.scatter(alphas, np.log10(g1_an), s=100, facecolors='none', \
            edgecolors='black', label='$g_1$ (analytical)')

# ax1.set_xlabel('')
ax1.set_ylabel('$\log_{10}$($g_1$) (yr$^{-1}$)')
ax1.set_title('Mode 1 frequency ($g_1$) vs.  (3-planet case)')
ax1.legend()

ax2.scatter(alphas, np.log10(g2_num), s=100, \
            marker='o', label='$g_2$ (numerical)')
ax2.scatter(alphas, np.log10(g2_an), s=100, facecolors='none', \
            edgecolors='black', label='$g_2$ (analytical)')
ax2.set_xlabel('')
ax2.set_ylabel('$\log_{10}$($g_2$) (yr$^{-1}$)')
ax2.set_title('Mode 2 frequency ($g_2$) vs.  (3-planet case)')
ax2.legend()

fig_path = "Figures/3p_mode_freqs.png"
plt.savefig(fig_path, bbox_inches='tight')
plt.show()
```

## Mode amplitudes

### Mode amplitude vectors

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(14,16))
fig.patch.set_facecolor('white')

ax1.scatter(alphas, v1_num[0], s=100, \
            marker='o', alpha=0.7, label='$A_{B1}$ (numerical)')
ax1.scatter(alphas, v1_num[1], s=100, \
            marker='o', alpha=0.7, label='$A_{C1}$ (numerical)')
ax1.scatter(alphas, v1_num[2], s=100, color='red', \
            marker='o', alpha=0.7, label='$A_{D1}$ (numerical)')

ax1.set_ylabel('Amplitude (deg)')
ax1.set_ylim(0, 4)
ax1.set_title('Numerically-derived mode 1 amplitudes (3-planet case)')
ax1.legend()

ax2.scatter(alphas, v2_num[0], s=100, \
            marker='o', alpha=0.7, label='$A_{B2}$ (numerical)')
ax2.scatter(alphas, v2_num[1], s=100, \
            marker='o', alpha=0.7, label='$A_{C2}$ (numerical)')
ax2.scatter(alphas, v2_num[2], s=100, color='red', \
            marker='o', alpha=0.7, label='$A_{D2}$ (numerical)')

ax2.set_xlabel('')
ax2.set_ylabel('Amplitude (deg)')
ax2.set_ylim(0, 2)
ax2.set_title('Numerically-derived mode 2 amplitudes (3-planet case)')
ax2.legend()

fig_path = "Figures/3p_mode_amps_num.png"
plt.savefig(fig_path, bbox_inches='tight')
plt.show()
```

### Relative differences between analyical and numerical results

```python
rel_amp_diffs_1 = v1_an - v1_num
rel_amp_diffs_2 = v2_an - v2_num
```

```python
## plot relative amp differences
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(16,18))
fig.patch.set_facecolor('white')

## mode 1
ax1.scatter(alphas, rel_amp_diffs_1[0], s=100, marker='x', \
            label='$A_{B1}$ difference')
ax1.scatter(alphas, rel_amp_diffs_1[1], s=100, marker='x', \
            label='$A_{C1}$ difference')
ax1.scatter(alphas, rel_amp_diffs_1[2], s=100, marker='x', \
            label='$A_{D1}$ difference')

#ax1.set_xlabel('')
ax1.set_ylabel('Amplitude (deg)')
#ax1.ticklabel_format(axis='y', style='sci', scilimits=(-2,-2))

ax1.set_ylim(0, 0.15)
ax1.set_title('Difference between analytical and numerical mode 1 amplitudes')
ax1.legend()

## mode 2
ax2.scatter(alphas, rel_amp_diffs_2[0], s=100, marker='x', \
            label='$A_{B2}$ difference')
ax2.scatter(alphas, rel_amp_diffs_2[1], s=100, marker='x', \
            label='$A_{C2}$ difference')
ax2.scatter(alphas, rel_amp_diffs_2[2], s=100, marker='x', \
            label='$A_{D2}$ difference')

ax2.set_xlabel('')
ax2.set_ylabel('Amplitude (deg)')
#ax2.ticklabel_format(axis='y', style='sci', scilimits=(-2,-2))

ax2.set_ylim(0, 0.5)
ax2.set_title('Difference between analytical and numerical mode 2 amplitudes')
ax2.legend()

fig_path = "Figures/3p_mode_amp_diffs.png"
plt.savefig(fig_path, bbox_inches='tight')
plt.show()
```
